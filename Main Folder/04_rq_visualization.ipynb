### Hotel Prices in Lisbon

```python
from utils import lisbon_hotels
df_h = lisbon_hotels()

# Replace NaN price levels with "-"
df_h["price_level"] = df_h["price_level"].fillna("-")

# Count hotels by price level
price_counts = df_h["price_level"].value_counts().reset_index()
price_counts.columns = ["price_level", "Count"]

# Plot
fig = px.bar(price_counts,
             x="price_level",
             y="Count",
             color="Count",
             color_continuous_scale="Oranges",
             title="Hotel Price Levels in Lisbon")

fig.update_traces(textposition="outside")
fig.update_layout(uniformtext_minsize=8, uniformtext_mode="hide")

fig.show()
```

Hotel prices in Lisbon have a great balance, showing a total of 21 expensive hotels and 20 mid-low/cheap hotels.

This is a good range to create packages such as luxury packages or budget packages for TripAdvisor customers.

This also shows that Lisbon hotels price range is good for packages that focus on other things rather then just the price, like ratings and reviews! 
Since price range is very balanced, when creating a package that focus on ratings and reviews it is safe to ignore the price levels and give the best package for customers who just look for quality.

### Heatmap

```python
df_r = lisbon_rest()

# Clean values
df_r["price_level"] = df_r["price_level"].fillna("-")
df_r = df_r[df_r["price_level"] != "-"]
df_r["rating"] = df_r["rating"].round(1)

pivot_counts = df_r.pivot_table(
    index="price_level",
    columns="rating",
    values="name",
    aggfunc="count",
    fill_value=0
)

# Adding Reviews to the heatmap
pivot_reviews = df_r.pivot_table(
    index="price_level",
    columns="rating",
    values="num_reviews",
    aggfunc="sum",   
    fill_value=0
)

customdata = np.expand_dims(pivot_reviews.values, axis=-1).tolist()

# Build heatmap
fig = px.imshow(
    pivot_counts,
    labels=dict(x="Rating", y="Price Level", color="Restaurant Count"),
    color_continuous_scale="Reds",
    title="Heatmap: Rating vs Price Level (Restaurants in Lisbon)",
    text_auto=True
)

# Hover Template
fig.update_traces(
    customdata=customdata,
    hovertemplate=(
        "<b>Price Level:</b> %{y}<br>"
        "<b>Rating:</b> %{x}<br>"
        "<b>Count:</b> %{z}<br>"
        "<b>Avg. Reviews:</b> %{customdata[0]:.1f}<extra></extra>"
    )
)

fig.update_layout(xaxis_side="top", height=600)
fig.show()
```

This heatmap proves our theory.
We have the majority of the restaurants sitting at low-medium prices, and, with the highest ratings and number of  reviews.
Biggest presence on the Heatmap being 11 restaurants at $$-$$$ price and with a rating of 4.8 and over a total of 
+18k reviews!! Which shows most tourists in Lisbon look for low-medium price restaurants and love it.
On the other hand we only have 7 restaurants on the high price category but two restaurants achieved a rating above 4 and more then 7.5k reviews.

Of course this is not 100% precise because not everyone reviews and rates restaurants but we can still get a general assumption of what tourists want when they come to eat in Lisbon.
